```python
import pandas as pd
import json
from pathlib import Path
```

```python
records = []
for p in Path("/mnt/backup/PeopleGatorNamedFaces/people-gator-named-faces-evaluation/report").glob("*.json"):
    *model_name, algorithm, ext = p.name.split(".")
    model_name = ".".join(model_name[1:])
    record = json.loads(p.read_text())
    record["model_name"] = model_name
    record["algorithm"] = algorithm
    records.append(record)
df = pd.DataFrame(records)
#expand metrics dict into columns
metrics_df = df["metrics"].apply(pd.Series)
df = pd.concat([df.drop(columns=["metrics", "confusion", "support"]), metrics_df], axis=1)
df.sort_values(["algorithm", "model_name"], inplace=True)
df.set_index(["algorithm", "model_name"], inplace=True)
algo_average = df.groupby("algorithm").mean()
model_best = df.groupby("model_name").mean()
df
```
